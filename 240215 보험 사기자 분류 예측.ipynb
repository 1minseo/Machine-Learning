### 개인소득 결측치 처리

```python
cust[cust.CUST_INCM.isnull()].head()
```

```python
occp_age_cust = cust.pivot_table(index=["AGE", "OCCP_GRP_1"], 
                                 values="CUST_INCM", aggfunc="mean")
occp_age_cust.head()

#pivot_table(): df -> 피벗 테이블로 변환
```

```python
occp_age_cust["CUST_INCM"] #피벗 테이블 occp_age_cust에서 CUST_INCM 열 선택. CUST_INCM 열에 해당하는 시리즈 객체 반환
```

```python
def fill_income(row): #cust의 각 행에 대해 고객 소득을 채워넣음
    try:
        avg_income = occp_age_cust["CUST_INCM"][row.AGE][row.OCCP_GRP_1]
    except:
        avg_income = 0
    return avg_income
```

```python
cust["CUST_INCM_NEW"] = None #null값만 따로 모아놓은
cust["CUST_INCM_NEW"][cust.CUST_INCM.isnull()] = cust[cust.CUST_INCM.isnull()].apply(fill_income, axis=1)

#CUST_INCM_NEW열을 None값으로 초기화
#apply() 함수에 fill_income함수를 적용하여 결측치를 채우고, 이를 CUST_INCM_NEW열에 할당
#CUST_INCM_NEW열에 CUST_INCM열의 결측치를 채운 결과가 저장됨
```

```python
missingno.bar(cust)
plt.show()
```

```python
cust["CUST_INCM"][cust.CUST_INCM.isnull()] = cust[cust.CUST_INCM.isnull()].apply(fill_income, axis=1).copy()
del cust["CUST_INCM_NEW"]

#copy() 메소드를 사용해 원본 df가 변경되지 않도록 복사본을 만듦
#결측치를 채운 결과가 CUST_INCM열에 저장되며, CUST_INCM_NEW열은 삭제됨
```

### DOC_SIU_RATIO : 의사별 사기비율

```python
# 의사별 사기수/전체 청구수
data = pd.merge(claim, cust[['CUST_ID', 'SIU_CUST_YN']], #왼쪽조인을 수행하여 claim df에 SIU_CUST_YN열 추가
                how='left', on='CUST_ID')
의사별전체 = data.pivot_table(index=['CHME_LICE_NO'], values='POLY_NO', #의사별 전체 청구 횟수 계산을 위해 pivot_table() 함수
                              aggfunc='count').reset_index()
의사별_YN = data.pivot_table(index=['CHME_LICE_NO', 'SIU_CUST_YN'], 
                             values='POLY_NO', 
                             aggfunc='count').reset_index()
의사별_YN = 의사별_YN[의사별_YN['SIU_CUST_YN']=='Y']
의사별전체 = pd.merge(의사별전체, 의사별_YN, how='left', 
                      on='CHME_LICE_NO')
의사별전체['DOC_SIU_RATIO'] = 의사별전체['POLY_NO_y']/의사별전체['POLY_NO_x'] #의사별 사기 비율 계산
의사별전체.fillna(0, inplace=True) #결측치를 0으로 채움
의사별사기비율 = 의사별전체[['CHME_LICE_NO', 'DOC_SIU_RATIO']]
의사별사기비율.head(5)
```

### SMOTE를 이용한 오버샘플링

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_resampled, y_resampled = sm.fit_resample(train_X, train_y)

#
#SMOTE(Synthetic Minority Over-sampling Technique): 소수 클래스의 샘플을 합성하여 데이터의 클래스 불균형을 완화
#증가된 특징 행렬을 X_resampled, 증가된 타겟 변수를 y_resampled로 저장
```

```python
# 1 기타 오버 샘플링 방식
from collections import Counter
from imblearn.over_sampling import BorderlineSMOTE
# OverSampling 방법(BSMOTE)
BSMOTE = BorderlineSMOTE()
# OverSampling 실행
X_resampled_BS, y_resampled_BS = BSMOTE.fit_resample(train_X, train_y)
# OverSampling 결과 데이터 확인
print("Before Sampling: ",Counter(train_y))
print("After oversampling(BSMOTE)): ",Counter(y_resampled_BS))

#BorderlineSMOTE: Borderline-SMOTE1과 Borderline-SMOTE2 두 가지 방법
#소수 클래스와 다수 클래스의 경계에 위치한 샘플을 중점적으로 샘플 생성 => SMOTE에 비해 효과적으로 클래스 불균형 해소
```

```python
# 2 기타 오버 샘플링 방식
from imblearn.over_sampling import ADASYN
# OverSampling 방법(ADASYN)
ADASYN = ADASYN()
# OverSampling 실행
X_train_ADASYN, y_train_ADASYN = ADASYN.fit_resample(train_X, train_y)
# OverSampling 결과 데이터 확인
print("Before Sampling: ",Counter(train_y))
print("After oversampling(ADASYN): ",Counter(y_train_ADASYN))

#ADASYN (Adaptive Synthetic Sampling): 데이터의 밀도에 따라 합성된 샘플을 생성
#소수 클래스 샘플에 대해서는 적은 수의 샘플 생성 (밀도에 따라 동적으로)
```

```python
# 3 기타 오버 샘플링 방식
from imblearn.over_sampling import SVMSMOTE
# OverSampling 방법(SVMSMOTE)
SVMSMOTE = SVMSMOTE()
X_train_SVMSMOTE, y_train_SVMSMOTE = SVMSMOTE.fit_resample(train_X, train_y)
print("Before Sampling: ",Counter(train_y))
print("After oversampling(ADASYN): ",Counter(y_train_SVMSMOTE))

#SVMSMOTE(Support Vector Machine SMOTE): SVM 분류기를 기반으로 생성
#SMOTE: 모든 소수 클래스 샘플에 대해 합성된 샘플을 생성
#SVMSMOTE: 소수 클래스와 다수 클래스의 경계에 있는 샘플에 집중하여 샘플을 생성

#모두 SMOTE의 단점인 무작위 샘플 합성의 한계를 극복하기 위해 개발. 데이터셋의 특성에 따라 적절한 방법을 선택
```

### MLPClassifier

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE #불균형한 클래스 분포를 보정하기 위해 SMOTE를 사용해 훈련 데이터를 증가시킴

def model_fit_assessment(X, y, model):
    sm = SMOTE()
    train_X, test_X, train_y, test_y = train_test_split(X, y, 
        test_size=0.3, stratify=y, shuffle=True, random_state=1)
    X_resampled, y_resampled = sm.fit_resample(train_X, train_y)
    model.fit(X_resampled, y_resampled)
    pred = model.predict(test_X)
    print(classification_report(test_y, pred))
```

```python
from sklearn.neural_network import MLPClassifier
mlp_model = MLPClassifier(hidden_layer_sizes=(30, 30, 20, 20))
model_fit_assessment(X, y, mlp_model)
```

### SVC (Support Vector Classifier)

```python
from sklearn.svm import SVC
svm_model = SVC(random_state=42)
model_fit_assessment(X, y, svm_model) #X: 특성 데이터, y: 타겟 데이터
```

### xgboost (Xtreme Gradient Boosting) 

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier(max_depth=10, learning_rate=0.01, n_estimators=100,
                          eval_metric='logloss', use_label_encoder=False)

model_fit_assessment(X, y, xgb_model)

#트리의 최대깊이=10, 학습률=0.01(트리의 가중치 조절), 생성할 트리의 개수=100, 모델을 평가할 지표=logloss, 라벨 인코더 사용 여부(카테고리형 변수 처리)
#logloss: 로그 손실. 다중 클래스 분류 문제에서 모델의 예측값과 실제값 간의 차이를 측정. 예측이 정확할수록 작아짐
```
